### **1. In how many cities Summer Olympics is held so far?**

**Method-1:**

```python
data = []

for City in df['City'].unique():
    data.append([City])
```

```python
data
```

```python
unique_cities=pd.DataFrame(data).count()
print(unique_cities)
```

### **2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
df2=df.loc[df['Medal']=='Gold']
```

```python
df2
```

```python
df2.groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot.bar(figsize = (5,5))
```

### **3. Which sport is having most number of medals so far? (Top 5)**

```python
df.groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot.bar(figsize = (5,5))
```

### **4. Which player has won most number of medals? (Top 5)**

```python
df.groupby('Athlete').count()['Medal'].sort_values(ascending = False).head().plot.bar(figsize = (5,5))
```

### **6. In which year India won first Gold Medal in Summer Olympics?**

```python
df6=df[(df['Country']=='IND') & (df['Medal']=='Gold')] 
df61=df6.sort_values(by='Year')
#print(df6.head(1))
print 'India won first Gold Medal in Summer Olympics in the year',
print df61.iloc[1][0]
```
